```python
import pandas as pd
```

```python

def get_env():
    """
    Returns environement AWS variables in dictionary
    """

    df = pd.read_table("~/.aws/credentials", delimiter="=").T
    secret_id = df['aws_access_key_id '].values[0].replace(' ', '')
    secret_key = df['aws_secret_access_key '].values[0].replace(' ', '')
    smart_api = df['smart_api_key '].values[0].replace(' ', '')

    env_dict = {
        "aws_access_key_id".upper(): secret_id,
        "aws_secret_access_key".upper(): secret_key,
        "aws_request_payer".upper(): "requester",
        "smart_api_key".upper(): smart_api

    }

    return env_dict
```

```python
my_secrets = get_env()
```

```python
import os
try:
    
        env_dict = get_env()
        os.environ["AWS_ACCESS_KEY_ID"] = env_dict["AWS_ACCESS_KEY_ID"]
        os.environ["AWS_SECRET_ACCESS_KEY"] = env_dict["AWS_SECRET_ACCESS_KEY"]
        os.environ["AWS_REQUEST_PAYER"] = env_dict["AWS_REQUEST_PAYER"]

except:
        import os
        os.environ["AWS_REQUEST_PAYER"] = 'requester'
```
