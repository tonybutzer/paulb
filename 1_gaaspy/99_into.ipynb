```python
help (Fmap)
```

```python
REGIONS='/iarpa/rylie/usgs-smart-rga/site_models'
```

```python
def return_map_please(geojson_file):
    f=Fmap(color='red')
    mymap = f.sat_geojson(f'{REGIONS}/{geojson_file}', geojson_file)
    return(mymap)
```

```python
from os import listdir
from os.path import isfile, join

def return_geojson_list():
    mypath=REGIONS
    onlyfiles = sorted([f for f in listdir(mypath) if isfile(join(mypath, f))])
    return(onlyfiles)
```

```python
import panel as pn
pn.extension()
```

```python
geojson_list = return_geojson_list()
print(geojson_list)
```

```python
title = '## Geojson Display Panel App'

geo_select = pn.widgets.Select(name='geo1', options=geojson_list)

@pn.depends(geo_select.param.value)
def refresh_geojson_map(geo1):
    return(return_map_please(geo_select.value))
```

```python
row = pn.Row(
    pn.Column(title, pn.panel(refresh_geojson_map, height=600, width=800), geo_select)
)

row.servable()
```
